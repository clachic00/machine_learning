```python
from sklearn.preprocessing import scale     # 데이터 스케일링
from sklearn.datasets import load_iris      # iris 데이터셋
from sklearn.cluster import KMeans        # K-Means 군집화 모델
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 폰트 및 마이너스 깨짐 설정
plt.rcParams['font.family'] = 'Malgun Gothic'   # 맑은 고딕
plt.rcParams['axes.unicode_minus'] = False      # 마이너스(-) 깨짐 방지

iris = load_iris()
```

```python
iris = load_iris()
iris_df = pd.DataFrame(
    data = iris.data,
    columns= iris.feature_names     # 특성 이름
)
iris_df.head()
```

```python
kmeans_model = KMeans(
    n_clusters=3,           # 군집 개수 (분류 클래스 수)
    init = 'k-means++',     # 초기 중심점 설정 방식
    max_iter=300,           # 최대 반복 횟수
    random_state=0
)
kmeans_model.fit(iris_df)   # K-Means 모델 학습
```

- init 옵션
    - kmeans++ (기본값, 가장 권장)
        - 첫 중심은 랜덤, 이후는 이미 선택된 중심과 멀리 떨어진 데이터를 우선 선택하여 중심점이 잘 퍼지도록 보장하는 방식
    - random
        - 데이터 포인트 중에서 무작위로 k개를 선택
        - 결과가 불안할 가능성이 있다.
    - 커스텀 방식 (고급)
        - 도메인마다 다르게 적용해준다.
        - 성능이 좋지 않을 가능성이 있다.


```python
kmeans_model.labels_    # K-Means가 학습한 결과. 어떤 군집에 속하는지 군집번호로 확인
```

```python
iris_df['cluster'] = kmeans_model.labels_       # K-Means로 할당된 군집 
iris_df['target'] = iris.target                 # 

iris_df
```

```python
iris_df['cluster'].value_counts()   # 군집별 데이터 갯수 확인 (실제 라벨과는 11개 차이)
```

```python
# 4차원 특성은 시각화가 안되므로 2차원으로 축소해서 본다.
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_transformed = pca.fit_transform(iris_df.iloc[:,0:4])    # 원본 4개 특성

pca_transformed
```

```python
iris_df['pca_x'] = pca_transformed[:,0]     # 첫번째 주성분 좌표
iris_df['pca_y'] = pca_transformed[:,1]     # 두번째 주성분 좌표

iris_df
```

```python
marker0 = iris_df[iris_df['cluster'] == 0].index # 군집 0에 속한 데이터의 연속된 수열
marker0
```

```python
marker1 = iris_df[iris_df['cluster'] == 1].index # 군집 0에 속한 데이터의 연속된 수열
marker1
```

```python
marker2 = iris_df[iris_df['cluster'] == 2].index # 군집 0에 속한 데이터의 연속된 수열
marker2
```

```python
# 산점도 그래프로 시각화
plt.scatter(
    x=iris_df.loc[marker0,'pca_x'],
    y=iris_df.loc[marker0,'pca_y'],
    marker='o'
) # 군집 0 산점도

# 산점도 그래프로 시각화
plt.scatter(
    x=iris_df.loc[marker1,'pca_x'],
    y=iris_df.loc[marker1,'pca_y'],
    marker='s'
) # 군집 1 산점도

# 산점도 그래프로 시각화
plt.scatter(
    x=iris_df.loc[marker2,'pca_x'],
    y=iris_df.loc[marker2,'pca_y'],
    marker='^'
) # 군집 2 산점도

plt.xlabel('PCA1')      # x축 이름 : 첫 번째 주성분
plt.ylabel('PCA2')      # y축 이름 : 두 번째 주성분
plt.show()
```

```python
# 2차원 데이터 생성해서 사용한다.
from sklearn.datasets import make_blobs

X,y = make_blobs(
    n_samples=200,          # 생성할 데이터의 갯수
    n_features=2,           # 특성의 갯수(2차원)
    centers=3,              # 군집 갯수
    cluster_std=0.9,        # 군집 퍼짐 정도
    shuffle=True,           # 데이터 섞음
    random_state=0
)

X
```

```python
cluster_df = pd.DataFrame(
    data=X,
    columns=['ftr1','ftr2']     # 컬럼 이름
)
cluster_df['target'] = y
cluster_df
```

```python
marker0 = cluster_df[cluster_df['target'] == 0].index # 실제 군집 0 인덱스
marker1 = cluster_df[cluster_df['target'] == 1].index # 실제 군집 1 인덱스
marker2 = cluster_df[cluster_df['target'] == 2].index # 실제 군집 2 인덱스

# 산점도 그래프로 시각화
plt.scatter(
    x=cluster_df.loc[marker0,'ftr1'],
    y=cluster_df.loc[marker0,'ftr2'],
    marker='o'
) # 군집 0 산점도

# 산점도 그래프로 시각화
plt.scatter(
    x=cluster_df.loc[marker1,'ftr1'],
    y=cluster_df.loc[marker1,'ftr2'],
    marker='s'
) # 군집 1 산점도

# 산점도 그래프로 시각화
plt.scatter(
    x=cluster_df.loc[marker2,'ftr1'],
    y=cluster_df.loc[marker2,'ftr2'],
    marker='^'
) # 군집 2 산점도

plt.xlabel('ftr1')      # x축 이름 : 첫 번째 주성분
plt.ylabel('ftr2')      # y축 이름 : 두 번째 주성분
plt.show()
```

```python
kmeans_model = KMeans(
    n_clusters=3,           # 군집갯수
    init='k-means++',       # 초기 중심점 설정 방식
    max_iter=300,           # 최대 반복 횟수
    random_state=0
)
kmeans_model.fit(
    cluster_df.drop('target',axis=1)
)   # 실제 모델은 제외하고 군집 학습

```

```python
kmeans_model.labels_    # 데이터가 할당된 군집 번호
```

```python
cluster_df['cluster'] = kmeans_model.labels_    # K-Means로 할당된 군집 번호 추가
cluster_df
```

```python
marker0 = cluster_df[cluster_df['cluster'] == 0].index # 실제 군집 0 인덱스
marker1 = cluster_df[cluster_df['cluster'] == 1].index # 실제 군집 1 인덱스
marker2 = cluster_df[cluster_df['cluster'] == 2].index # 실제 군집 2 인덱스

# 산점도 그래프로 시각화
plt.scatter(
    x=cluster_df.loc[marker0,'ftr1'],
    y=cluster_df.loc[marker0,'ftr2'],
    marker='o'
) # 군집 0 산점도

# 산점도 그래프로 시각화
plt.scatter(
    x=cluster_df.loc[marker1,'ftr1'],
    y=cluster_df.loc[marker1,'ftr2'],
    marker='s'
) # 군집 1 산점도

# 산점도 그래프로 시각화
plt.scatter(
    x=cluster_df.loc[marker2,'ftr1'],
    y=cluster_df.loc[marker2,'ftr2'],
    marker='^'
) # 군집 2 산점도

plt.xlabel('ftr1')      # x축 이름 : 첫 번째 주성분
plt.ylabel('ftr2')      # y축 이름 : 두 번째 주성분
plt.show()
```
