```python
import pandas as pd
import numpy as np

movies = pd.read_csv('data/movies.csv')
ratings = pd.read_csv('data/ratings.csv')
```

```python
movies
```

```python
ratings
```

```python
# movieId 기준으로 평점 데이터와 영화 정보 데이터를 병합
rating_movie = pd.merge(ratings,movies,on='movieId')
rating_movie
```

```python
# 피벗테이블 생성 (인덱스 : 아이템, 컬럼 : 유지)
rating_matirix = rating_movie.pivot_table(
    index="title",          # 행 : 영화 제목
    columns='userId',       # 열 : 사용자 ID
    values='rating'         # 값 : 평점
).fillna(0)

rating_matirix
```

```python
from sklearn.metrics.pairwise import cosine_similarity # cosine 유사도 계산 도구
item_sim = cosine_similarity(
    rating_matirix,
    rating_matirix
)       # 영화간에 유사도 계산 (아이템 - 아이템끼리)

```

```python
rating_matirix.index
```

```python
item_sim_df = pd.DataFrame(
    item_sim,
    index = rating_matirix.index,
    columns= rating_matirix.index
) 
item_sim_df
```

```python
item_sim_df.index
```

```python
item_sim_df['Toy Story (1995)'].sort_values(ascending=False)
```

```python
# 영화 추천해주는 함수
def movie_recommendations(movie_name,n):return item_sim_df[movie_name].sort_values(ascending=False)[1:n+1].index
```
