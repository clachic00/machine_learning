```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python

movies = pd.read_csv('data/movies.csv')
ratings = pd.read_csv('data/ratings.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
movie_ratings = pd.merge(ratings, movies, on="movieId")
movie_ratings
```

```python
title_user = movie_ratings.pivot_table(
    index='userId',             # 행 : userId
    columns='title',            # 열 : title
    values='rating'             # 값 : 벌점
).fillna(0)                     # 벎점을 안준 것은 0으로 처리
title_user
```

```python
# cosine 유사도 계산
user_based_result = cosine_similarity(title_user)
user_based_result
```

```python
# 데이터프레임으로 만든다.
user_based_df = pd.DataFrame(user_based_result,index=title_user.index, columns=title_user.index)
user_based_df
```

```python
# 정렬시켜서 결과 10개 확인하기
user_based_df.iloc[0].sort_values(ascending=False).head(10)
```

```python
title_user.iloc[0].index    # 0번 유저 (userId = 1)의 모든 영화리스트 뽑아낸다.
```

```python
mylist = []
for index, temp in enumerate(title_user.iloc[0]):
    if temp > 0:
        print(title_user.iloc[0].index)
        print("=======================================================")
        print(title_user.iloc[0].index[index])
        print("=======================================================")
        mylist.append(title_user.iloc[0].index[index])
        
```

```python
u_data = title_user.loc[266]    # 유저아이디 266인 사람의 데이터
u_data[u_data >= 4].index   # 평점이 4점 이상인 영화제목
```

```python
user_title = []
for temp in [266, 311, 368, 57, 91]:
    u_data = title_user.loc[temp]
    titles = u_data[u_data >= 4].index
    user_title.extend(titles)               # 중복제거하면서 추가된다.

user_title      # 나와 가장 유사한 5명의 추천리스트
```

```python
# 나와 가장 비슷한 5명이 추천한 영화 리스트 - 내가 본 영화 리스트 = 추천리스트
set(user_title) - set(mylist)
```
